Problem -2
Plot the scores of 2 students in 5 different subjects.Subjects as x-axis,marks as y-axis

```python
stu1=[98,90,92,95,85]
stu2=[90,89,91,50,45]
stu3=[89,76,54,84,90]
Sub=["Maths","Science","English","Telugu","Social"]
plt.plot(Sub,stu1,label="student1")
plt.plot(Sub,stu2,label="student2")
plt.plot(Sub,stu3,label="student3")
plt.legend()
```

```python
stu1=[98,90,92,95,85]
stu2=[90,89,91,50,45]
Sub=["Maths","Science","English","Telugu","Social"]
plt.subplot(2,1,1)
plt.plot(Sub,stu1,label="student1",color='b')
plt.legend()
plt.subplot(2,1,2)
plt.plot(Sub,stu2,label="student2",color='g')
plt.legend()
```

Create Subplots(bar) showing the profits of two companies
Given:
Year   A   B
2019  230 200
2020  560 160
2021  780 270
2022  127 127
2023  128 400

```python
compA=np.array([230,560,780,127,128])
compB=np.array([200,160,270,127,400])
print("Revenues of company A",compA)
print("Revenues of company B",compB)
profA=np.diff(compA)
print("profits of A",profA)
profB=np.diff(compB)
print("profits of B",profB)
years=["19-20","20-21","21-22","22-23"]
plt.subplot(1,2,1)
plt.bar(years,profA,label="Profit A",color="k")
plt.legend(loc = "best")
plt.subplot(1,2,2)
plt.bar(years,profB,label="Profit B",color="g")
plt.legend(loc="best")


```
